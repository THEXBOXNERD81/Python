### Plotting in Matplotlib can be done in either of two ways, which ones? Which way is the recommended approach?

By either using the pyplot function or the object-oriented way and doing some more complex manipulation on the plot. When doing more easier and fast programing, pyplot i prefered. Doing the object-oriented way is when doing more bigger and complex program that require more complexity.


### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

Figure: Top level Artist, which holds all plot elements. Can be seen as a canvas
Axes: An Axes object encapsulates all the elements of an individual (sub-)plot in a figure.
Axis: Represents a plot with lines, ticks and labels.
Artist: An artist is the element that is visible on the figure.

### When plotting in Matplotlib, what is the expected input data type?

It is expected that a numpy.array or objects that can be passed as numpy.asarray as input, otherwise the plot may not be as intended.

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
x = np.linspace(-4, 4)

fig, ax = plt.subplots(figsize=(5, 2.7), layout='constrained')
ax.plot(x, x**2, label='quadratic')
ax.set_xlabel('Bottom Text')
ax.set_ylabel('Side Text')
ax.set_title('Top Text')
ax.legend()
```

### Create a figure containing 2  subplots where the first is a scatter plot and the second is a bar plot. You have the data below. 

```python
# Data for scatter plot
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)
x = np.linspace(-2, 2, 100)
y = x**2

# Data for bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(7, 3), layout='constrained')
ax1.scatter(random_data_x, random_data_y)
ax1.set_xlabel('Bottom Text')
ax1.set_ylabel('Side Text')
ax1.set_title('Top text')
ax2.bar(names, values)
ax2.set_title('Top Text')
ax2.set_xlabel('Bottom Text')
```
